```python
# Filter by'country'(USA)
df_states = df[(df["country"] == "United States") & (df["state"].notnull())]
df_states.head()
```

Filtering the rows of the Dask DataFrame (df) to include only those rows where the state column is not empty and the nation column has the value "United States"

# 3.2

```python
df_2 = df_states[(df_states["date"] >= "2020-01-01") & (df_states["date"] <= "2021-02-28")]
df_2.head()
```

Further filters the dataset to include only data within the time period January 1, 2020, to February 28, 2021.

```python
morrs = df_2.groupby("state").agg(
    {"deaths": "sum", "population": "mean"}
).compute()
morrs["per_capita_mortality"] = (
    morrs["deaths"] / morrs["population"]
)

# Rank states
morrs['rank'] = morrs['per_capita_mortality'].rank(ascending=False)

print(morrs.sort_values(by='rank'))
```

Determines the average population, adds up all of the deaths, and then groups the data by state to determine the per-capita mortality for each state. The ratio of all deaths to the average population for each state is then used to calculate the per-capita mortality. Per capita mortality is then used to rank the states, with the greatest mortality ranking first.

# 3.3

Assumption 1: The likelihood of detecting cases and deaths is consistent over the course of the outbreak.

Assumption 2: All detected cases have resolved (that is, reported cases have either recovered or died).

```python
# Extract month from the date
df_3 = df[(df["date"] >= "2020-01-01") & (df["date"] <= "2021-02-28")]
df_3 = df_3[(df_3["country"] == "United States") & (df_3["state"].notnull())]
df_3["month"] = df_3["date"].dt.to_period("M")

# Calculate the total deaths and cases per state and month
mcfr = df_3.groupby(["state", "month"]).agg({
    "cases": "sum",
    "deaths": "sum"
}).compute()

# Calculate the CFR
mcfr["CFR"] = (mcfr["deaths"] / mcfr["cases"]) * 100

all_months = pd.period_range("2020-01", "2021-02", freq="M")

# Reindex
cfr_matrix = mcfr["CFR"].unstack().reindex(columns=all_months, fill_value=0)

# Display the CFR matrix (50 states x 14 months)
print(cfr_matrix)
```

Same process as 3.2 but this time we are calculating Case fatality ratio (CFR) which is an proportion of individuals diagnosed with a disease who die from that disease and is therefore a measure of severity among detected cases.

![WHO](https://www.who.int/images/default-source/health-topics/coronavirus/cfr.png?sfvrsn=c5031cde_1)

Finally we are pivoting data to get the needed matrix to further explore the dataset.





# 3.4

```python
# Calculate month-to-month changes in CFR for each state
cfr_changes = cfr_matrix.diff(axis=1)  # Calculate the difference across months

# Sum the changes to see how much the CFR increased or decreased over time for each state
total_changes = cfr_changes.abs().sum(axis=1)

ranked_states = total_changes.rank(ascending=False)

# Display the ranking of states based on CFR changes
print(ranked_states.sort_values())
```

We obtain monthly variations in each state's Case Fatality Rate (CFR), add up these variations by adding up their absolute values, and then rank the states according to the overall size of the variations. The findings are shown to indicate which states saw the biggest changes in CFR over time.
